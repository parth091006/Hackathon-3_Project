```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px

# 1. Page Configuration
st.set_page_config(page_title="My First Dashboard", layout="wide")

st.title("📊 Business Analytics Dashboard")
st.markdown("Welcome to your new Streamlit dashboard!")

# 2. Sidebar for Filters
st.sidebar.header("Filter Options")
category = st.sidebar.multiselect(
    "Select Category:",
    options=["Sales", "Marketing", "R&D"],
    default=["Sales", "Marketing"]
)

# 3. Generating Mock Data (Replace this with your CSV/SQL load)
chart_data = pd.DataFrame(
    np.random.randn(20, 3),
    columns=['Sales', 'Marketing', 'R&D']
)

# 4. Layout: Metrics
col1, col2, col3 = st.columns(3)
col1.metric("Revenue", "$50,000", "+5%")
col2.metric("Active Users", "1,200", "-2%")
col3.metric("Churn Rate", "0.5%", "+0.1%")

st.divider()

# 5. Layout: Charts
left_column, right_column = st.columns(2)

with left_column:
    st.subheader("Performance over Time")
    st.line_chart(chart_data[category])

with right_column:
    st.subheader("Category Distribution")
    # Using Plotly for more interactivity
    fig = px.pie(values=[40, 30, 30], names=["Sales", "Marketing", "R&D"])
    st.plotly_chart(fig, use_container_width=True)
```
